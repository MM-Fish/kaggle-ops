## アンサンブル

```python
import pandas as pd
import numpy as np

file_name_list = ['_LightAutoML_with_FE-4.867', 'moving-median-ps4.935']
run_name_list = ['_LightAutoML_with_FE', 'moving-median']
len_subs = len(file_name_list)
subs = []
target = 'congestion'
for i in range(len_subs):
  file_name = file_name_list[i]
  run_name = run_name_list[i]
  subs.append(pd.read_csv(f'../models/{file_name}/{run_name}_submission.csv'))
```

```python
submission = subs[0][['row_id']]
submission[target] = round(pd.concat(subs, axis=1)[target].median(axis=1))

submission.to_csv('../models/_ensemble/submission.csv', index=False)
```
